# ë‹µì„ ì•Œê³  ë§ì·„ì„ í™•ë¥ 
> ê°ê´€ì‹ ë¬¸ì œë¥¼ í‘¸ëŠ” ê³¼ì •ì—ì„œ, í•™ìƒì€ ë‹µì„ ì´ë¯¸ ì•Œê³  ìˆê±°ë‚˜ ì°ì–´ì•¼ í•©ë‹ˆë‹¤.
í•™ìƒì´ ë‹µì„ ì•Œ í™•ë¥ ì€  ğ‘ ì´ê³ , ë‹µì„ ì°ì–´ì„œ ë§ì¶œ í™•ë¥ ì´  $1\over{m}$  ì´ë¼ê³  í•  ë•Œ  ğ‘š ê³¼  ğ‘ ë¥¼ ì…ë ¥ë°›ì•„ í•™ìƒì´ ë¬¸ì œì˜ ë‹µì„ ì•Œê³  ë§ì·„ì„ í™•ë¥ ì„ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”.

## ì¶”ë¡ ê³¼ì •
- P(ì•Œ|ì •) : ì •ë‹µì¸ë°, ë‹µì„ ì•Œê³  ë§ì·„ì„ í™•ë¥ 
- P(ì•Œ) = $p$
- P(ëª°) = 1 - $p$
- P(ì •|ì•Œ) : ì•Œì•„ì„œ ì •ë‹µì¼ í™•ë¥  = 1
- P(ì •|ëª°) : ëª¨ë¥´ëŠ”ë° ì°ì–´ì„œ ë§ì¶œ í™•ë¥  = $1\over{m}$
- P(ì•Œ|ì •) = ${P(ì •|ì•Œ) P(ì•Œ)} \over {P(ì •|ì•Œ) P(ì•Œ) + P(ì •|ëª°) P(ëª°)}$

```python
def correct(p, m):
    # (1) * (p) / (1) * (p) + (1 / m) * (1 - p)
    # p / p + (1 / m) * (1 - p)
    ans = (p * m) / (p * (m - 1) + 1)
    return ans
    
correct(0.25, 4)
```

# ì–‘ì„±íŒì • ì‹œ, ì‹¤ì œë¡œ ì§ˆë³‘ì„ ê°€ì§€ê³  ìˆì„ í™•ë¥ 
> íŠ¹ì • ì§ˆë³‘ì„ ê°€ì§€ê³  ìˆëŠ” ê²½ìš° 99% ì˜ í™•ë¥ ë¡œ íƒì§€ í•  ìˆ˜ ìˆëŠ” ì‹¤í—˜ë°©ë²•ì´ ìˆìŠµë‹ˆë‹¤.
ê·¸ëŸ¬ë‚˜ ë™ì‹œì— ì´ ë°©ë²•ì€, 1% ì˜ í™•ë¥ ë¡œ ì§ˆë³‘ì´ ì—†ì§€ë§Œ ì§ˆë³‘ì´ ìˆë‹¤ê³  ì§„ë‹¨ í•˜ëŠ” ê²½ìš°ë„ ìˆëŠ”ë°ìš”.
ì‹¤ì œë¡œ ëª¨ë“  ì¸êµ¬ì¤‘ 0.5% ë§Œì´ ì´ ì§ˆë³‘ì„ ê°€ì§€ê³  ìˆë‹¤ê³  í•˜ê² ìŠµë‹ˆë‹¤.
íŠ¹ì • ì‚¬ëŒì´ ì´ ë°©ë²•ì„ í†µí•´ ì§ˆë³‘ì´ ìˆë‹¤ê³  ì§„ë‹¨ ë°›ì•˜ì„ë•Œ, ì‹¤ì œë¡œ ì§ˆë³‘ì„ ê°€ì§€ê³  ìˆì„ í™•ë¥ ì„ êµ¬í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”.

## ì¶”ë¡ ê³¼ì •
- P(tpr|+) : ì§ˆë³‘ì´ ìˆë‹¤ê³  ì§„ë‹¨ ë°›ì•˜ì„ë•Œ, ì‹¤ì œë¡œ ì§ˆë³‘ì„ ê°€ì§€ê³  ìˆì„ í™•ë¥ 
- P(tpr) : ë°œë³‘ë¥  = prior = 0.005
- P(fpr) : ì§ˆë³‘ì—†ìŒ = 1 - prior = 0.995
- P(+|tpr) : ê±¸ë ¸ëŠ”ë° ì–‘ì„± = tpr = 0.99
- P(+|fpr) : ì•ˆê±¸ë ¸ëŠ”ë° ì–‘ì„± = 1 - tpr = fpr = 0.01 
- p(tpr|+) = ${p(+|tpr) p(tpr)} \over {p(+|tpr) p(tpr) + p(+|fpr)p(fpr)}$ 

```python
def disease(prior, tpr, fpr):
    ans = (tpr * prior) / ((tpr * prior) + (fpr * (1-prior)))
    return ans
    
disease(0.005, 0.99, 0.001)
```

- P(ìœ ì£„) = ans1

- P(ìƒˆì¦ê±°|ë¬´ì£„) : ë¬´ì£„ì´ë©´ì„œ ìƒˆì¦ê±°ì˜ íŠ¹ì§•ì„ ê°€ì¡Œì„ í™•ë¥  = 0.1
