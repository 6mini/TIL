### Feature Engineering

Relative Perfomance 라는 새로운 feature를 계산하세요.

이는 최근 1년치 매출액의 평균값을 기준으로 

- 10% 이상 -> `S`
- 5% 이상 -> `A`
- -5 ~ 5% -> `B`
- -5%이하 -> `C`
- 10%이하 -> `D`

라는 값을 갖는 feature입니다.

20년도 2분기에 해당하는 결과값은 `A`가 나와야합니다.

각각에 해당하는 등급이 나오기 위해서 필요한 매출액을 추가로 서술하세요.

```python
s = (sum(df['매출액']) / 5) + (sum(df['매출액']) / 5 * 0.1)
a = (sum(df['매출액']) / 5) + (sum(df['매출액']) / 5 * 0.05)
c = (sum(df['매출액']) / 5) - (sum(df['매출액']) / 5 * 0.05)
d = (sum(df['매출액']) / 5) - (sum(df['매출액']) / 5 * 0.1)
print('S등급 :', s, '이상')
print('A등급 :', a, '이상')
print('B등급 :', c, '~', a)
print('C등급 :', c, '이하')
print('D등급 :', d, '이하')
```

```python
def RP(mechoorack):
    if mechoorack >= s:
        return 'S'
    elif mechoorack >= a:
        return 'A'
    elif a > mechoorack > c:
        return 'B'
    elif mechoorack <= c:
        return 'C'
    else:
        return 'D'
```

```python
df['Relative Perfomance'] = df['매출액'].apply(RP)
df
```
