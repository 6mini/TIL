### Feature Engineering

Relative Perfomance ë¼ëŠ” ìƒˆë¡œìš´ featureë¥¼ ê³„ì‚°í•˜ì„¸ìš”.

ì´ëŠ” ìµœê·¼ 1ë…„ì¹˜ ë§¤ì¶œì•¡ì˜ í‰ê· ê°’ì„ ê¸°ì¤€ìœ¼ë¡œ 

- 10% ì´ìƒ -> `S`
- 5% ì´ìƒ -> `A`
- -5 ~ 5% -> `B`
- -5%ì´í•˜ -> `C`
- 10%ì´í•˜ -> `D`

ë¼ëŠ” ê°’ì„ ê°–ëŠ” featureì…ë‹ˆë‹¤.

20ë…„ë„ 2ë¶„ê¸°ì— í•´ë‹¹í•˜ëŠ” ê²°ê³¼ê°’ì€ `A`ê°€ ë‚˜ì™€ì•¼í•©ë‹ˆë‹¤.

ê°ê°ì— í•´ë‹¹í•˜ëŠ” ë“±ê¸‰ì´ ë‚˜ì˜¤ê¸° ìœ„í•´ì„œ í•„ìš”í•œ ë§¤ì¶œì•¡ì„ ì¶”ê°€ë¡œ ì„œìˆ í•˜ì„¸ìš”.

```python
s = (sum(df['ë§¤ì¶œì•¡']) / 5) + (sum(df['ë§¤ì¶œì•¡']) / 5 * 0.1)
a = (sum(df['ë§¤ì¶œì•¡']) / 5) + (sum(df['ë§¤ì¶œì•¡']) / 5 * 0.05)
c = (sum(df['ë§¤ì¶œì•¡']) / 5) - (sum(df['ë§¤ì¶œì•¡']) / 5 * 0.05)
d = (sum(df['ë§¤ì¶œì•¡']) / 5) - (sum(df['ë§¤ì¶œì•¡']) / 5 * 0.1)
print('Së“±ê¸‰ :', s, 'ì´ìƒ')
print('Aë“±ê¸‰ :', a, 'ì´ìƒ')
print('Bë“±ê¸‰ :', c, '~', a)
print('Cë“±ê¸‰ :', c, 'ì´í•˜')
print('Dë“±ê¸‰ :', d, 'ì´í•˜')
```

```python
def RP(mechoorack):
    if mechoorack >= s:
        return 'S'
    elif mechoorack >= a:
        return 'A'
    elif a > mechoorack > c:
        return 'B'
    elif mechoorack <= c:
        return 'C'
    else:
        return 'D'
```

```python
df['Relative Perfomance'] = df['ë§¤ì¶œì•¡'].apply(RP)
df
```
